```python
# movies per director 
# retunrs adictionary of movie counts per director 
def director_movie_count():
    movie_dir = dict()
    dirs = w_df['director']
    for d in dirs:
        dn = d.split(":")
        n = dn[1].strip()
        if n in movie_dir:
            movie_dir[n] += 1
        else: 
            movie_dir[n] = 1 
    return movie_dir
movie_count_by_dir = director_movie_count()
```

```python
movie_rating_by_year = w_df.groupby(['Year', 'title', 'IMDbRating']).size().reset_index()
```

```python
def actor_movie_count():
    movie_actors = dict()
    actors = w_df['stars']
    for actor in actors:
        an = actor.split('\n')[2].split(',')
        n = an[1].strip()
        if n in movie_actors:
            movie_actors[n] += 1
        else: 
            movie_actors[n] = 1 
    return movie_actors
movie_count_by_actor = actor_movie_count()
```
